```python
connection = http.client.HTTPConnection('api.football-data.org')
headers = { 'X-Auth-Token': '65330cec3e6346a3be018b7067091650', 'X-Response-Control': 'minified' }
connection.request('GET', '/v1/competitions/426/fixtures', None, headers )
response = json.loads(connection.getresponse().read().decode())
#print(json.dumps(response, indent=4, sort_keys=True)) #pring json in a nice format
#print(response["fixtures"][1])
#print(len(response['fixtures']))
#fd={}
#for resp in response['fixtures']:
#    fd["date"]=resp['date']
#    fd["MatchNumber"]=resp['matchday']
#print(fd)
result=[]
for resp in response['fixtures']:
    result.append(str(datetime.strptime(resp['date'], '%Y-%m-%dT%H:%M:%SZ').strftime("%Y-%m-%d")) + "," + 
                   str(resp['matchday'])+ ","+ str(resp['homeTeamName']) + "," + str(resp['awayTeamName']) + "," + str(resp['result']['goalsHomeTeam']) + "-" + 
                   str(resp['result']['goalsAwayTeam']))
    
#print(result)
#with open("EPL/2016-2017.csv", 'w', newline='', encoding='utf-8') as csvfile:
#    writer = csv.writer(csvfile)
#    writer.writerows(result)
print(result)
#file_name = ticker_name + '_new.csv'
soccer_file = open("EPL/2016-2017.csv", 'w')
for line in result:
    soccer_file.write(line)
    soccer_file.write('\n')
soccer_file.close()
```
